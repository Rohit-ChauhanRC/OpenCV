# Goals
#### 1. How to read image, how to display it and how to save it back
#### 2. In this session we learn : cv2.imread(), cv2.imshow(), cv2.imwrite()
#### 3. We learn how to display images with Matplotlib

## Loading dependencies :

```python
import numpy as np
import cv2
```

```python
# Load an color image in grayscale
img = cv2.imread("images/img1.jpg",0) # we can simply pass intergers 1, 0, -1.
```

```python
img
```

```python
# Display an image
cv2.imshow('display_image', img) # First argument is a window name which is a string.Second argument is our image
cv2.waitKey(0)  # It is akeyboard binding function.
cv2.destroyAllWindows() # Simply desttroys all windows we created.
```

## write an image
#### Use the function cv2.imwrite() to save an image

```python
cv2.imwrite('My_gray.png',img) # This will save the image in PNG format in the working directory.
```

## Using Matplotlib

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img1 = cv2.imread('images/img1.jpg',0)
plt.imshow(img1, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # To hide tick values on X and Y axis
plt.show()
```
